## 2. Filtering out

```python
FILTER_OUT_KEYWORDS = [
    r'xxx',
    r'sex',
    r'porn',
    r'fuck',
    r'tube',
    r'love\.',
    r'xnxx',
    r'love',
    r'pussy',
    r'girlfriend',
    r'znakom',
]
def is_abult_content_domain(domain, example_url):
    """return true if the given domain contain adult content keywords"""    
    
    for keyword in FILTER_OUT_KEYWORDS:
        if re.search(keyword, domain):
            return True

        
    return False
```

```python

def test_is_abult_content_domain():
    print(is_abult_content_domain('xxxporn.com', 'xxxporn.com/yXcDsc3DOa1') == True)
    print(is_abult_content_domain('webhirsche.at ', 'webhirsche.at/th/team/ ') == False)
    
test_is_abult_content_domain()
    
```

## 3. Estimate number of pages (URLs) to scrape from all domains in `domains_thai_fitered`

```python
# domains_thai_fitered
```

```python
with open('../data/en-de.bicleaner07.v2.json', 'r', encoding='utf-8') as f:
    
    domains_urls_list = json.load(f)
    
    
```

```python
number_of_pages = Counter()
domains_thai_fitered_names = list(map(lambda x: x[0], domains_thai_fitered))
for domain, urls in domains_urls_list['de'].items():
    
    if domain in domains_thai_fitered_names:
        number_of_pages[domain] += len(urls)
```

```python
sorted(number_of_pages.items(), key=lambda x: -x[1])
```

```python
print('total number of URLs (estimated): {}'.format(sum(number_of_pages.values())))
```

```python
def get_number_of_pages(given_domain):
    for domain, urls in domains_urls_list['de'].items():
        if domain == given_domain:
            return len(urls)
        
    return 0
```

```python
def print_domain_and_urls(dataset):
    
    for (domain, data) in dataset:
        url_th = data['example_modified_url'][0]
        url_en = url_th.replace('th', 'en')
        n = get_number_of_pages(domain)
        print('domain: {}\n\texample_th: {}\n\texample_en: {}\n\tnumber_of_pages: {}\n'.format(domain, url_th, url_en, n))
```

```python
print_domain_and_urls(domains_thai_fitered)
```
